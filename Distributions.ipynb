```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv("C:\\Users\\alackey\\Downloads\\Distributions.csv")
df
```

```python
#Starting with VitD_levels, first create histogram with kde.
col = 'VitD_levels'

data = df[col]
mean = data.mean() #Determine Mean
median = data.median() #Determine Median

std = data.std()
plus_one_std = mean + std
minus_one_std = mean - std

fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(data, ax = ax, stat='probability', kde=True)
ax.axvline(mean, color = 'red') #Display Mean line in red.
ax.axvline(median, color = 'green') #Display Median line in green.
ax.axvline(plus_one_std, color = 'black')
ax.axvline(minus_one_std, color = 'black') #Std -1/+1 range in black.
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0) #Span fill

ax.set_title(col)
```

```python
kurt = stats.kurtosis(data, fisher = False)
kurt
```

**1. Is it discrete or continuous?**

Continuous

**2. Does it have a skew? If so, which direction? (+/-)**

It appears symmetrical

**3. What type of kurtosis does it display? (Meso, Lepto, Platy)**

Kurt of ~3 means Mesokurtic.
